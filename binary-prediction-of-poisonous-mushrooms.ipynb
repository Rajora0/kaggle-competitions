```python
import numpy
import pandas 
import pycaret
import sklearn

print(numpy.__version__)
print(pandas.__version__)
print(pycaret.__version__)
print(sklearn.__version__)
```

```python
import pandas as pd

from pycaret.classification import *

from sklearn.model_selection import train_test_split
```

```python
train_file_path = "../input/playground-series-s4e8/train.csv"
data = pd.read_csv(train_file_path)

df_train, df_test = train_test_split(data.sample(n=100), test_size=.2)
```

```python
s = setup(df_train, target = 'class',session_id = 123, ignore_features='id')
```

```python
best = compare_models(sort='MCC')
```

```python
predict_model(best, data = df_test)
```

```python
s = setup(data, target = 'class',session_id = 123, ignore_features='id')
rf = create_model('rf')
tuned_rf = tune_model(rf)
```

```python
test_file_path = "../input/playground-series-s4e8/test.csv"
test_data = pd.read_csv(test_file_path)
test_data
```

```python
predictions = predict_model(tuned_rf, data = test_data)
predictions
```

```python
sample_submission_df = predictions[["id","prediction_label"]].rename(columns={"prediction_label":"class"})
sample_submission_df.to_csv('/kaggle/working/submission.csv', index=False)
sample_submission_df.head()
```
